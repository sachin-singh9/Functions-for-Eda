#### 3. Often, we are not required to plot the graphs for all the variables in our dataset. Add an additional argument to the function in 2 named ‘var’. This will take a list containing the variable index and return the graphs for only those variables. By default, (i.e. if the list of variable index is not provided), then it must return the required graphs for all the variables in data. 
Example:
Graphs(data, var=c(1,3,4))

-> Will generate the graphics for only the variables 1,3 & 4 in the data.

```python
#Q3
def Graphs(data,var):
    
    
    data_slice=data.iloc[:,var]
    
    numeric_data = data_slice.select_dtypes(include=[np.number])
    categorical_data = data_slice.select_dtypes(exclude=[np.number])
    n=numeric_data.columns
    c=categorical_data.columns
    
    for i in list(n):
        
        fig = plt.figure(figsize=(10,10))
        
        sns.distplot( numeric_data[i] ,kde=False)
        fig.savefig(i+'_Hist.png')
    for k in list(n):   
        fig = plt.figure(figsize=(10,8))
        sns.boxplot(numeric_data[k])
        fig.savefig(k+'_Box.png')
            
        
       
    for j in list(c):
        fig = plt.figure(figsize=(10,8))
        sns.countplot(x=categorical_data[j],data=categorical_data)
        fig.savefig(j+'_Bar.png')
        
```

```python
Graphs(cars,var=[1,2,8])
```

#### 4. Sometimes, we do not want to mess up our working directories with so many image files. Create an additional argument for the function "dir" (directory), such that the function exports all the files to that specified folder (which need not necessarily be your working directory). (10 marks)
Example:
Graphs(data, Variable=c(1,3,4), dir=".../Praxis/Graphs")

-> will generate all the necessary graphics for the variables 1, 3 and 4 in the specified location in your system which is ".../Praxis/ Graphs"

```python
#Q4
def Graphs(data,var,dir):
    
    
    data_slice=data.iloc[:,var]
    
    numeric_data = data_slice.select_dtypes(include=[np.number])
    categorical_data = data_slice.select_dtypes(exclude=[np.number])
    n=numeric_data.columns
    c=categorical_data.columns
    
    for i in list(n):
        fig,axis=plt.subplots(1,2)
        fig = plt.figure(figsize=(10,10))
        
        sns.distplot( numeric_data[i], kde=False,ax=axis[0])
        sns.boxplot(numeric_data[i],ax=axis[1])
        plt.savefig(dir+ i+'.png')
        plt.show()
    for j in list(c):
        fig = plt.figure(figsize=(10,8))
        sns.countplot(x=categorical_data[j],data=categorical_data)
        plt.savefig(dir+j+'.png')
        plt.show()
```
